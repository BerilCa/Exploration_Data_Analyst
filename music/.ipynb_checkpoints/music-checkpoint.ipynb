### Analisis Perilaku Mendengarkan Musik

```python
stream = df['Primary streaming service'].value_counts()

plt.pie(stream, labels=stream.index, autopct='%1.1f%%')
plt.show()
```

```python
variables = ['Age', 'Hours per day', 'Anxiety', 'Depression', 'Insomnia', 'OCD']

num_cols = 2
num_rows = (len(variables) + 1) // num_cols

fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 8))
axes = axes.flatten()

for i, var in enumerate(variables):
    axes[i].hist(df[var], bins=10, color='skyblue', edgecolor='gray')
    axes[i].set_title(var)
    
if len(variables) < num_cols * num_rows:
    axes[len(variables)].axis('off')
    
plt.tight_layout()
plt.show()
```

```python
variables = ['Frequency [Classical]',
       'Frequency [Country]', 'Frequency [EDM]', 'Frequency [Folk]',
       'Frequency [Gospel]', 'Frequency [Hip hop]', 'Frequency [Jazz]',
       'Frequency [K pop]', 'Frequency [Latin]', 'Frequency [Lofi]',
       'Frequency [Metal]', 'Frequency [Pop]', 'Frequency [R&B]',
       'Frequency [Rap]', 'Frequency [Rock]', 'Frequency [Video game music]']

num_cols = 4
num_rows = (len(variables) + 1) // num_cols

fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 8))
axes = axes.flatten()

for i, var in enumerate(variables):
    axes[i].hist(df[var], bins=10, color='green', edgecolor='gray')
    axes[i].set_title(var)
   
    
if len(variables) < num_cols * num_rows:
    axes[len(variables)].axis('off')
    
plt.tight_layout()
plt.show()
```

```python
df['Anxiety'].mean()
```

```python
# Memilih variabel yang ingin dianalisis
variables = ['Anxiety', 'Depression', 'Insomnia', 'OCD']
genre_variable = 'Fav genre'

# Mengelompokkan data berdasarkan genre musik favorit dan menghitung rata-rata variabel-variabel terkait mental health
grouped_data = df.groupby(genre_variable)[variables].mean()

# Menampilkan visualisasi
grouped_data.plot(kind='bar', figsize=(10, 6))
plt.title('Mean Mental Health Scores by Favorite Music Genre')
plt.xlabel('Favorite Music Genre')
plt.ylabel('Mean Score')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.show()
```

```python
# Mengelompokkan data berdasarkan genre musik favorit
grouped = df.groupby('Fav genre')

# Menghitung rata-rata kesehatan mental di setiap grup
mental_health_stats = grouped[['Anxiety', 'Depression', 'Insomnia', 'OCD']].mean()

# Menampilkan statistik kesehatan mental berdasarkan genre musik favorit
print(mental_health_stats)
```

```python
fav = df['Fav genre'].value_counts()

plt.pie(fav, labels=fav.index, autopct='%1.1f%%')
plt.show()
```

```python
mental_health_vars = ['Anxiety', 'Depression', 'Insomnia', 'OCD']
mental_health_stats = df[mental_health_vars].describe()
print(mental_health_stats)
```

```python
eff = df['Music effects'].value_counts()

plt.pie(eff, labels=eff.index, autopct='%1.1f%%')
plt.show()
```
