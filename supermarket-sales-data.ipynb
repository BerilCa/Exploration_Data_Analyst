# 1 Descriptive Statistics

**1.1 Transaction Data with Item Category**

```python
df = pd.merge(df_name, df_sales, on='item_code')

df.head()
```

**1.2 Count of Items in Each Category with Non-negative Quantity Sold.**

```python
df_min = df[df['qty_sold(kg)'] < 0]
df_min
```

```python
filtered = df[df['qty_sold(kg)'] >= 0 ]
no_minus_cat = filtered.groupby(['category_code', 'category_name']).size().reset_index(name='no_minus')
no_minus_cat.sort_values(by='no_minus', ascending=False, inplace=True)

no_minus_cat
```

**1.3 Summary Statistics of Quantity Sold for Each Item Category**

```python
filtered = df[df['qty_sold(kg)'] >= 0 ]
statistics_sold = filtered.groupby(['category_code', 'category_name']).agg(
    total_sold=('qty_sold(kg)', 'sum'),
    max_of_qty=('qty_sold(kg)', 'max'),
    min_of_qty=('qty_sold(kg)', 'min'),
    avg_of_qty=('qty_sold(kg)', 'mean')).reset_index()

statistics_sold
```

**1.4 Summary Statistics of Total Price Sold for Each Item Category**

```python
filtered = df[df['qty_sold(kg)'] >= 0]
filtered['total_price'] = filtered['qty_sold(kg)'] * filtered['unit_selling(rmb_kg)']

statistics_price = filtered.groupby(['category_code', 'category_name']).agg(
    total_price=('total_price', 'sum'),
    max_price=('total_price', 'max'),
    min_price=('total_price', 'min'),
    avg_price=('total_price', 'mean')
).round(2).reset_index()


filtered.drop(columns=['total_price'], inplace=True)

statistics_price
```

**1.5 Summary Statistics of Unit Selling Prices for Each Item Category**

```python
filtered = df[df['qty_sold(kg)'] >= 0 ]
statistics_unit_selling = filtered.groupby(['category_code', 'category_name']).agg(
    total_unit_selling=('unit_selling(rmb_kg)', 'sum'),
    max_unit_selling=('unit_selling(rmb_kg)', 'max'),
    min_unit_selling=('unit_selling(rmb_kg)', 'min'),
    avg_unit_selling=('unit_selling(rmb_kg)', 'mean')).round(2).reset_index()

statistics_unit_selling
```

**1.6 Summary Statistics of Sales Returns by Item Category**

```python
filtered = df[df['qty_sold(kg)'] < 0]

filtered['total_price'] = filtered['qty_sold(kg)'] * filtered['unit_selling(rmb_kg)']

# Melakukan pengelompokan dan menghitung statistik
sale_return = filtered.groupby(['category_code', 'category_name']).agg(
    no_of_return=('sale/return', lambda x: sum(x == 'return')),
    sum_of_qty=('qty_sold(kg)', 'sum'),
    max_of_qty=('qty_sold(kg)', 'max'),
    min_of_qty=('qty_sold(kg)', 'min'),
    avg_of_qty=('qty_sold(kg)', 'mean'),
    sum_of_unit_px=('unit_selling(rmb_kg)', 'sum'),
    max_of_unit_px=('unit_selling(rmb_kg)', 'max'),
    min_of_unit_px=('unit_selling(rmb_kg)', 'min'),
    avg_of_unit_px=('unit_selling(rmb_kg)', 'mean'),
    sum_of_price=('total_price', 'sum'),
    max_of_price=('total_price', 'max'),
    min_of_price=('total_price', 'min'),
    avg_of_price=('total_price', 'mean')
).round(2).reset_index()

# Menampilkan hasil
sale_return
```

# **2. Calculate Item Profit Margin**

```python
#merger data frame df_sales,df_price_per_kg,df_name
sales_wholesale = pd.merge(df_sales, df_price_per_kg, how = 'left', on = ['item_code', "date"])
sales_wholesale.head()
```

```python
df_wls_cat = pd.merge(sales_wholesale, df_name, how = 'left', on = ['item_code'])
df_wls_cat.head()
```

```python
df_wls_cat['whole_sale_price'] = df_wls_cat['qty_sold(kg)'] * df_wls_cat['wholesale_price(rmb_kg)']
df_wls_cat['selling_price'] = df_wls_cat['unit_selling(rmb_kg)'] * df_wls_cat['qty_sold(kg)']

profit_margin = df_wls_cat.groupby(['item_name']).agg(
    avg_wholesale =('wholesale_price(rmb_kg)', 'mean'),
    max_wholesale =('wholesale_price(rmb_kg)', 'max'),
    min_wholesale =('wholesale_price(rmb_kg)', 'min'),
    whole_sale_price =('whole_sale_price', 'sum'),
    selling_price = ('selling_price', 'sum')
    
).round(2).reset_index()
profit_margin['profit']=profit_margin['selling_price'] - profit_margin['whole_sale_price']

profit_margin
```
